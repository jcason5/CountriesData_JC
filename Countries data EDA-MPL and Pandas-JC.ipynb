#### Q2. What is the distribution of life expectancy across all countries and years?


```python
plt.hist(countries['lifeExpectancy'],facecolor="skyblue", edgecolor="black", bins=25, range=(15,90))
plt.xlabel("Life Expectancy")
plt.ylabel("Frequency")
plt.title("Distribution of Life Expectancy")
plt.show()
```

#### Q3. How does life expectancy vary over the years for a specific country (e.g., Afghanistan)?
https://www.geeksforgeeks.org/matplotlib-markers-module-in-python/
https://www.geeksforgeeks.org/grids-in-matplotlib/


```python
Afghanistan_data=countries[countries['country']=='Afghanistan']
Afghanistan_data.head()
```

```python
group_data=Afghanistan_data.groupby('year')['lifeExpectancy'].median()

plt.plot(group_data, color='green', marker='o')

plt.title("Life Expectancy over the years in Afghanistan")
plt.xlabel('Year')
plt.ylabel('Life Expectancy')

plt.grid()
plt.show()
```

#### Q4. What is the relationship between GDP per capita and life expectancy?


```python
plt.scatter(countries['gdpPerCapita'], countries['lifeExpectancy'], facecolor='red')
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")
plt.grid()
plt.show()
```

#### Q5. How has the population changed over time for a specific continent (e.g., Africa)?

** Use groupby for year with populations

```python
Africa_data=countries[countries['continent']=='Africa']
```

```python
group_data=Africa_data.groupby('year')['population'].sum()

plt.plot(group_data, color='red', marker='o')

plt.title("Population over Time in Africa")
plt.xlabel('Year')
plt.ylabel('Population')

plt.grid()
plt.show()
```

#### Q6. What is the trend of GDP per capita over the years for a specific continent (e.g., Europe)?


```python
Europe_data=countries[countries['continent']=='Europe']
```

```python
group_data=Europe_data.groupby('year')['gdpPerCapita'].mean()

plt.plot(group_data, color='blue', marker='o')

plt.title("GDP per Capita Over Time in Europe")
plt.xlabel('Year')
plt.ylabel('GDP Per Capita')

plt.grid()
plt.show()
```
